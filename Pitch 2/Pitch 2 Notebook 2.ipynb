```python
import pandas as pd
```

```python
bkzips = pd.read_csv("brooklyn_zips.csv")
bkzips.info()
```

```python
givetake = pd.read_csv("donor_spend_zip.csv")
givetake.info()
```

```python
merged_df = pd.merge(bkzips, givetake, on="zip" )
merged_df
```

```python
#multiply amount - spending column by 1000 to create the true number; original data was in thousands
merged_df["amount - spending"] = 1000 * merged_df["amount - spending"]
merged_df
```

```python
#change column names to Donations, Capital Spending
merged_df = merged_df.rename(columns ={"amount - donors":"donations", "amount - spending":"capital spending"})
merged_df
```

```python
merged_df.sort_values("capital spending", ascending = False)
```

```python
merged_df.sort_values("donations", ascending = False)
```

```python
#what's the ratio of capital spending to donations across Brooklyn?
merged_df["ratio"] = merged_df["capital spending"]/merged_df["donations"]
merged_df
```

```python
#which zip has the highest ratio?
merged_df.sort_values("ratio", ascending = False)
```

```python
#divide capital spending by 1000 then find the ratio to donation
merged_df["capital spending"] = merged_df["capital spending"]/1000
merged_df
#suppress scientific notation
#pd.options.display.float_format = "{:.0f}".format
```

```python
#find ratio between new "per 1000 capital spending" with donations
merged_df["per 1000 ratio"] = merged_df["capital spending"]/merged_df["donations"]
merged_df
```

```python
merged_df.sort_values("per 1000 ratio", ascending = False)
```
